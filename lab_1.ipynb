```python
import math
import numpy as np
import gurobipy as gp
from gurobipy import GRB, quicksum as qsum
```

```python
k = 6  # k - номер студента по списку
l = 1  # l - номер группы

n = 10 + math.floor(k/4.0)
m = 30 - math.floor(k/4.0)

b = [70 + l + k + i for i in range(1, m + 1)]
c = [50 + l + k - i for i in range(1, n + 1)]
A = [
        [
            1 + ((j+k)*i + j*j + i*i*i + 3*(i+l)) % (30 + math.floor(k/5.0))
            for j in range(1, n + 1)
        ]
        for i in range(1, m + 1)
    ]

A = np.array(A)
b = np.array(b)
c = np.array(c)
```

```python
c
```

```python
try:
    model = gp.Model("lab_1_1")
    
    x = model.addMVar(n, lb=[0.0 for _ in range(n)],
                      vtype=GRB.CONTINUOUS, name="x")
    model.update()
    
    model.setObjective(c.T @ x, GRB.MAXIMIZE)
    model.addConstr(A @ x <= b)
    
    model.optimize()

    for v in model.getVars():
        print('%s = %g' % (v.VarName, v.X))

    print('Obj: %g' % model.ObjVal)

except gp.GurobiError as e:
    print('Error code ' + str(e.errno) + ': ' + str(e))

except AttributeError:
    print('Encountered an attribute error')
```
